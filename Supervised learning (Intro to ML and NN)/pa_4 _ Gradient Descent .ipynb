## Gradient descent for linear regression
## Without scaling

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from numpy.linalg import norm
```

```python
def transform_x(x, d):
    return np.column_stack([x**i for i in range(d+1)])
```

```python
def generate_points_polynomial(xmin, xmax, num_points, sigma,  beta_star, d_star):
    x_vec = np.random.uniform(xmin, xmax, num_points)
    X = transform_x(x_vec, d_star)
    y_star = np.random.normal(X.dot(beta_star), sigma)
    return x_vec, y_star
```

```python
def cost_function(beta, X, y,lam):
    reg = lam*norm(beta, ord = 2)
    cost = sum((1/(len(y)))*(beta@X.T - y)**2 + reg)
    return cost
```

```python
def regression_gradient_descent (X, y, lam = 0.0001, alpha=0.01, max_epoch = 1000, max_cost_before_break = 1000):
    times = {}
    beta = np.zeros(X.shape[1])
    prev_cost = float("inf")
    curr_cost = cost_function(beta, X, y, lam)
    eps = 0.0001
    cnt = 0
    L = {}
    while True :#np.abs(prev_cost - curr_cost) >= eps and cnt <= max_epoch :
        cost = cost_function(beta, X,y,lam)
        L[cnt] = curr_cost
        end = time.time()
        if int(end-begin) not in times:
            times[int(end-begin)] = cnt
            print ("%s seconds has passed at count %s out of %s" % (end-begin, cnt, max_count))
        if cost> max_cost_before_break:
            break
        if cnt>max_count:
            break
#         if np.abs(prev_cost - curr_cost) < eps:
#             break
        prev_cost = curr_cost
        beta -= alpha*(1/len(y))*(beta@X.T - y)@X + lam*2*beta
        curr_cost = cost_function(beta, X, y, lam)
        cnt += 1
    return beta, L
```

```python
xmin = -15
xmax = 20
d_star = 1
d_train = 1
beta_star = [1, 1, 0.025, - 0.005]
sigma = 2
num_points = 10000
lam = 0
max_count = 10000
x, y = generate_points_polynomial(xmin, xmax, num_points, sigma,  [1,1], d_star)
X = transform_x(x,d_train)
alpha_list = [0.01,0.001, 0.0001]


for alpha in alpha_list:
    begin = time.time()
    print ("Starting gradient descent for alpha=%s" % alpha)
   
    
    beta_grad, costs = regression_gradient_descent(X,y,lam, alpha,max_count)
    end = time.time()
    print ("Execution time =%s" % (end-begin))
    print ("Beta_grad COST is=%s\n" % cost_function(beta_grad,X,y,lam))
    plt.figure(1)

    plt.loglog(costs.keys(),costs.values(), label="alpha=%s and best_cost=%.2lf" % (alpha, min(costs.values())))
    plt.figure(2)

    plt.semilogy(costs.keys(),costs.values(), label="alpha=%s and best_cost=%.2lf" % (alpha, min(costs.values())))
plt.figure(1)
plt.legend()
plt.title("Learning curve, NO scaling,loglog, d_star=%s, sigma=%s,\n \
d_train=%s, numpoints=%s, starting beta = ZEROS" % (d_star,sigma,d_train, num_points))

plt.xlabel("Number of steps")
plt.ylabel("Cost Function")
plt.figure(2)
plt.title("Learning curve, NO scaling,Semilog, d_star=%s, sigma=%s,\n \
d_train=%s, numpoints=%s, starting beta = ZEROS" % (d_star,sigma,d_train, num_points))

plt.xlabel("Number of steps")
plt.ylabel("Cost Function")
plt.legend()
plt.show()
```
